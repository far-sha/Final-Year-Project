```python
import pandas as pd
import numpy
numpy.random.seed(2)
import matplotlib.pyplot as plt
laptop = pd.read_csv('laptop 3.csv')
laptop.head()
```

```python
y=laptop.Ram_GB
x=laptop.Cpu
```

```python
train_x = x[:80]
train_y = y[:80]

test_x = x[80:]
test_y = y[80:]

print("shape of original dataset :", laptop.shape)
print("shape of input - training set", train_x.shape)
print("shape of output - training set", train_y.shape)
print("shape of input - testing set", test_x.shape)
print("shape of output - testing set", test_y.shape)
```

```python
plt.scatter(train_x, train_y)
plt.scatter(test_x, test_y)
plt.xlabel('Cpu') 
plt.ylabel('Ram_GB')
plt.ylim([0, 70])
plt.rcParams["figure.autolayout"] = True
plt.show()
```

```python
plt.scatter(test_x, test_y)
plt.xlabel('Cpu') 
plt.ylabel('Ram_GB')
plt.ylim([0, 70])
plt.show()
```

```python
#polynomial training
mymodel1 = numpy.poly1d(numpy.polyfit(train_x, train_y, 4))

myline1 = numpy.linspace(0, 16, 100)

plt.scatter(train_x, train_y)
plt.plot(myline1, mymodel1(myline1))
plt.xlabel('Ram_GB') 
plt.ylabel('Price_RM')
plt.ylim([0, 30000])
plt.show()
```

```python
##polynomial testing
mymodel = numpy.poly1d(numpy.polyfit(test_x, test_y, 4))

myline = numpy.linspace(0, 70, 100)

plt.scatter(test_x, test_y)
plt.plot(myline, mymodel(myline))
plt.xlabel('Ram_GB') 
plt.ylabel('Price_RM')
plt.ylim([0, 30000])
plt.show()
```

```python
#R-squared test
r2 = r2_score(test_y, mymodel(test_x))

print(r2) #57%
```
