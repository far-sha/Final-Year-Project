```python
import pandas as pd
from sklearn.model_selection import train_test_split
laptop = pd.read_csv('laptop 2.csv')
laptop.head()
```

```python
y=laptop.Ram_GB
x=laptop.drop('Price_RM',axis=1)
```

```python
x.head()
```

```python
y.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```
