```python
import pandas as pd
import numpy
numpy.random.seed(2)
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
laptop = pd.read_csv('laptop 3.csv')
laptop.head()
```

```python
y=laptop.Ram_GB
x=laptop.Cpu

#train_x = x[:80]
#train_y = y[:80]

#test_x = x[80:]
#test_y = y[80:]

train_x,test_x,train_y,test_y = train_test_split(x,y,test_size=0.2)
print("shape of original dataset :", laptop.shape)
print("shape of input - training set", train_x.shape)
print("shape of output - training set", train_y.shape)
print("shape of input - testing set", test_x.shape)
print("shape of output - testing set", test_y.shape)
```

```python
plt.scatter(train_x, train_y, color='green')
plt.scatter(test_x, test_y, color='purple')
plt.xlabel('Cpu') 
plt.ylabel('Ram_GB')
plt.ylim([0, 70])

#input
p = ["Intel Core i5 2.3GHz"]
q = [16]
plt.plot(p, q, marker="o",markerfacecolor="red")

plt.rcParams["figure.autolayout"] = True
plt.show()
```
